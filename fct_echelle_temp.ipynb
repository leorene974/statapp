chacal met dans df ton df dans pas ton pas et dans l'echelle "min" ou "sec" bv chacal

```python
def echelle(df,pas,echelle):
    if echelle=="min":
        df.loc[:,"date"]=df.loc[:,"date"].str[11:16]
    elif echelle =="sec":
        df.loc[:,"date"]=df.loc[:,"date"].str[11:19]
    else :
        return ("echelle = min ou echelle = sec restart")
    
    df = df.drop_duplicates(subset="date").reset_index(drop=True)
    df=df.iloc[::pas,:]
    return df
```
